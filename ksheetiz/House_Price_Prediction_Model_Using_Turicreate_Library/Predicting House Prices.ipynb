# Import Turicreate

```python
import turicreate as tc
```

# Load some house sales data

```python
sales = tc.SFrame('house_data.sframe/')
```

```python
sales
```

# Exploring the data for house sales

```python
sales.show()
```

```python
tc.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

# Create a simple regression model of sqft_living to price

```python
train_data, test_data = sales.random_split(.8,seed = 0)
```

## Build the regression Model

```python
sqft_model = tc.linear_regression.create(train_data, target = 'price', features = ['sqft_living'])
```

# Evaluate the simple model

```python
print(test_data['price'].mean())
```

```python
print(sqft_model.evaluate(test_data))
```

# Let's show what our predictions look like

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_data['sqft_living'], test_data['price'],'.'
        test_data['sqft_living'],sqft_model.predict(test_data),'.')
```

```python
sqft_model.coefficients
```

# Build a regression model with more features

```python
my_features_model = tc.linear_regression.create(train_data,target = 'price', features = my_features)
```

```python
print(my_features)
```

```python
print(sqft_model.evaluate(test_data))
print(my_features_model.evaluate(test_data))
```

```python
print("New features added to my features ",my_features)
```

```python
updated_feature_model = tc.linear_regression.create(train_data,target='price',features = my_features)
```

```python
print(sqft_model.evaluate(test_data))
print(my_features_model.evaluate(test_data))
print(updated_feature_model.evaluate(test_data))
```
